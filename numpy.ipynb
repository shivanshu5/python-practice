array operations

```python
#arithmatic
a1 = np.array([1,2,3,4,5])
a2 = np.array([6,7,8,9,10])
a1 + a2 #add
a1 - a2 #substract

a1 * a2 #multiply
a1 / a2 #division
a1 // a2 #floorDivision
a1 ** a2 #Exponential
```
